```python
# Questions
# 1. Which food item has maximum sodium content? (target less than 2,000 milligrams/day)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as stat
from warnings import filterwarnings
```

```python
### Categorical scatterplots
```

```python
sns.set(style='whitegrid')
plot = sns.swarmplot(x="Category", y='Sodium', data=df, palette="Set2", size=3.6, marker="o")
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title('Sodium Content')
plt.show()
```

```python
# Understanding the summary statistics
# The maximum value of sodium is 3600
df['Sodium'].describe()
```

```python
# Compute the index values at which the maximum value of sodium is obtained in the data frame
df['Sodium'].idxmax()
```

```python
# Find the item assosiated with the 82nd item in the data frame
# The Chicken McNuggets (40 pieces) are the food item on the menu that has a highest sodium content.

df.at[82, 'Item']
```

```python
# Data exploration using visualizations
```

```python
#Using scatterplot with Protein on the x-axis and Total Fat on the y-axis
plot = sns.jointplot(x ="Protein", y ='Total Fat', data =df)
plot.show()
```

```python
# Find the Pearson correlation coefficient and the significance of the correlation (P)
df = pd.DataFrame(data, columns = ['Protein','Total Fat'])
#define the columns to perform calculations on
pearson_coef, p_value = stats.pearsonr(df["Protein"], df["Total Fat"]) 
print("Pearson Correlation Coefficient: ", pearson_coef, "and a P-value of:", p_value)
# The results shows that the variables are certainly correlated
```
